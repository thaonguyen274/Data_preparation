```python
amz = pd.read_csv("amzn.csv")
fb = pd.read_csv("fb.csv")
goog = pd.read_csv("goog.csv")
nflx = pd.read_csv("nflx.csv")
aapl = pd.read_csv("aapl.csv")
```

```python
amz = amz.assign(ticker = 'AMZ')
fb = fb.assign(ticker = 'FB')
goog = goog.assign(ticker = 'GOOG')
nflx = nflx.assign(ticker = 'NFLX')
aapl = aapl.assign(ticker = 'AAPL')
```

```python
faang = pd.concat([amz,fb,goog,nflx,aapl])
print(faang.shape)
faang.head()
```

```python
faang.to_csv('faang.csv')
```

```python
data = pd.read_csv("faang.csv", index_col=[0])
data.head()
```

```python
faang = faang.assign(
    date=pd.to_datetime(faang.date),
    volume=faang.volume.astype(int)
).sort_values(
    ['date', 'ticker']
)
```

# The European Centre for Disease Prevention and Control (ECDC) provides an open dataset on COVID-19 cases called daily number of new reported cases of COVID-19 by country worldwide. This dataset is updated daily, but we will use a snapshot that contains data from January 1, 2020 through September 18, 2020. Clean and pivot the data so that it is in wide format:

a)Read in the covid19_cases.csv file.

```python
covid = pd.read_csv('/content/drive/MyDrive/DATA PREPARATION /2. Pandas/Data/covid19_cases.csv')
covid.head()
```

```python
covid = covid.assign(date=lambda x: pd.to_datetime(x.dateRep, format='%d/%m/%Y')).set_index('date')
covid
```

```python
covid = covid. replace('United_States_of_America', 'USA').replace('United_Kingdom', 'UK').sort_index()
covid[covid.countriesAndTerritories =="USA"].head()
```

```python
covid[covid.countriesAndTerritories =="UK"].head()
```

```python
covid = covid.loc[covid['countriesAndTerritories'].isin(["Argentina", "Brazil", "China", "Colombia", "India", "Italy", "Mexico", "Peru", "Russia", "Spain", "Turkey", "UK",  "USA"])]
print(covid.shape)
covid.head()
```

```python
pivot_covid = covid.pivot(columns="countriesAndTerritories",values="cases").fillna(0)
pivot_covid.head()
```
